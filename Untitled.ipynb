```python
def count_tags_in_column(df, column_name, tags_to_count):  
    tag_counts = {}  
    for tag in tags_to_count:  
        # 使用str.contains()检查每个标签是否在列中的每个元素中出现  
        # 这里假设标签之间没有重叠，且大小写敏感  
        # 如果需要大小写不敏感，可以在str.contains中传入case=False  
        count = df[column_name].str.contains(tag, na=False).sum()  
        tag_counts[tag] = count  
    return tag_counts  
  
# 示例标签列表  
tags_to_count = ['华语', '流行', '治愈', '经典','欧美','日语','韩语','粤语','摇滚','民谣','电子','舞曲','说唱','轻音乐','爵士','乡村','RBSoul','古典','民族','英伦','金属','朋克','蓝调','雷鬼','世界音乐','拉丁','New Age','古风','后摇','BossaNova','清晨','夜晚','学习','工作','午休','下午茶','地铁','驾车','运动','旅行','散步','酒吧','怀旧','清新','浪漫','伤感','放松','孤独','感动','兴奋','快乐','安静','思念','综艺','影视原声','ACG','儿童','校园','游戏','70后','80后','网络歌曲','KTV','翻唱','吉他','钢琴','器乐','榜单','00后']  # 添加你的其他标签  
  
# 调用函数并打印结果  
tag_counts = count_tags_in_column(df, 'tagsum', tags_to_count)  

#在本地存为csv文件 方便进行运算
f = open('tags.csv', 'w', encoding='utf-8',newline='')
csv_writer = csv.writer(f)
csv_writer.writerow(['tag', 'count'])

for tag, count in tag_counts.items():  
    
    f = open('tags.csv','a',encoding='utf-8',newline='')

    csv_writer = csv.writer(f)

    csv_writer.writerow([tag,count])
    
    print(f"标签 '{tag}' 出现了 {count} 次")
```

```python
#TF-IDF挖掘歌单简介的关键字
brief = df_new['briefsum']
brief_list = list(brief)
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(brief)

df = pd.DataFrame(
    tfidf_matrix.toarray(),
    columns = vectorizer.get_feature_names_out()
)
top_words = {}
for i,document in enumerate(brief):
    top_indices = df.iloc[i].argsort()[-3:][::-1].values
    top_words[document] = [
        vectorizer.get_feature_names_out()[index]
        for index in top_indices
    ]
for doc,top in top_words.items():
    print(f"Document:'{doc}'")
    print(f"Top words:{','.join(top)}")
    print("------------------------")
```

```python
#用snownlp对标题做情感分析

from snownlp import SnowNLP
 
def sentiment_analysis(title):
    # 使用SnowNLP对中文文本进行情感分析
    s = SnowNLP(title)
    # SnowNLP的sentiments方法返回情感倾向分数，越接近1表明情感越积极，越接近0表明情感越消极
    sentiment_score = s.sentiments
    return sentiment_score

title_sen_list = []

for title in df_new['titlesum']:
    title_sen = sentiment_analysis(title)
    title_sen_list.append(title_sen)
    print(f"标题：{title},情感指数：{title_sen}")
```

```python
sen_sogood = 0
sen_good = 0 
sen_sobad = 0
sen_bad = 0 

for i in range(0,len(title_sen_list)-1):
    if title_sen_list[i] >= 0.75:
        sen_sogood = sen_sogood + 1
    elif 0.5 <= title_sen_list[i] < 0.75:
        sen_good = sen_good +1 
    elif title_sen_list[i] < 0.25:
        sen_sobad = sen_sobad + 1
    else:
        sen_bad = sen_bad + 1

print(sen_sogood)
print(sen_good)
print(sen_sobad)
print(sen_bad)
```

```python
cut_word_list = list(map(lambda x: ''.join(x), filtered_words2))
content_str = ' '.join(cut_word_list).split()
word_fre = pd.Series(_flatten(content_str)).value_counts()  # 统计词频
word_fre[:50]
```

```python
keywords = word_fre[:50].index
keywords
```

```python
matrix = np.zeros((len(keywords)+1)*(len(keywords)+1)).reshape(len(keywords)+1, len(keywords)+1).astype(str)
matrix[0][0] = np.NaN
matrix[1:, 0] = matrix[0, 1:] = keywords
matrix
```

```python
cont_list = [cont.split() for cont in cut_word_list]
for i, w1 in enumerate(word_fre[:50].index):
    for j, w2 in enumerate(word_fre[:50].index):
        count = 0
        for cont in cont_list:
            if w1 in cont and w2 in cont:
                if abs(cont.index(w1)-cont.index(w2)) == 0 or abs(cont.index(w1)-cont.index(w2)) == 1:
                    count += 1
        matrix[i+1][j+1] = count
```

```python
kwdata = pd.DataFrame(data=matrix)
kwdata.to_csv('关键词共现矩阵.csv', index=False, header=None, encoding='utf-8-sig')
```
